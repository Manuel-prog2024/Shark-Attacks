```python
# Assuming shark_df is already defined and contains the ‘Sex’ column with string values
# Step 1: Define the function
def replace_starting_string(text):
    if pd.isnull(text):  # Handle None or NaN values
        return text
    text_str = str(text)  # Convert to string to ensure proper handling
    if re.match(r"^(f|F)", text_str):
        return "F"
    elif re.match(r"^(m|M)", text_str):
        return "M"
    else:
        return text
# Step 2: Apply the function to the ‘Sex’ column
shark_df["Sex"] = shark_df["Sex"].apply(lambda x: replace_starting_string(x))
# Step 3: Get the value counts
new_shark_df = shark_df["Sex"].value_counts()
# Step 4: Filter the counts to only include values occurring at least 10 times
filtered_values = new_shark_df[new_shark_df >= 10].index
# Step 5: Mask and update the original ‘Sex’ column based on these filtered values
shark_df["Sex"] = shark_df["Sex"].apply(lambda x: x if x in filtered_values else None)
# Display the updated dataframe
display(shark_df.head())
display(filtered_df)
```
